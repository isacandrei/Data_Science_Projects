```python
import nltk
import matplotlib.pyplot as plt
from nltk.corpus import state_union

texts = state_union.fileids()
for text in texts:
    
    x=[]
    y=[]
    z=[]
    
    print ("\n\n"+text)
    words = [w for w in state_union.words(text)]

    n = len(state_union.words(text))
    
    frequencies = nltk.FreqDist([w for w in words])
    for e in range(0, len(frequencies.most_common(n))):
        word = frequencies.most_common(n)[e][0]
        x.append(word)
        y.append(frequencies.freq(word))
        
    for i in range(0, len(y)):
        z.append((1/(i+1)) * y[0])
        
    plt.plot(range(1, len(frequencies.most_common(n)) +1), y, 'r', label='TF')
    plt.plot(range(1, len(frequencies.most_common(n)) +1), z, 'b', label='Zipf')
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel('Rank of the words')
    plt.ylabel('Frequency of the words')
    plt.legend(loc='upper right')
    plt.show()
```

```python
import nltk
import string
from nltk.corpus import state_union
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

texts = state_union.fileids()
for text in texts:
    print ("\n\n"+text)
    
    # Stop words removal
    stop_words = set(stopwords.words('english'))
    
    clean_text = []
    
    for w in state_union.words(text):
        if w not in stop_words:
            clean_text.append(w)
    
    # Special Characters removal
    punctuations = list(string.punctuation)
    punctuations.append('--')
    punctuations.append('\'.')
    punctuations.append('.)')
    punctuations.append('...')
    
    # PorterStemmer
    clean_text2 = []
    
    for i in clean_text:
        p = PorterStemmer().stem(i)
        clean_text2.append(p)
        
    # Lemmatizer
    clean_text3 = []
    
    for q in clean_text2:
        x = WordNetLemmatizer().lemmatize(q)
        clean_text3.append(x)
    
    words = [w for w in clean_text3]

    frequencies = nltk.FreqDist([w for w in words if w not in punctuations])
    for e in frequencies.most_common(10):
        print(e)
```
